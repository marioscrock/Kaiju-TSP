```python
import matplotlib.pyplot as pyplot
import numpy as np
```

```python
timing  = np.genfromtxt('exp_csv/thriftTiming.csv', delimiter=',', dtype='str')
batches = np.array(timing[1:,0])
t_before = np.array(timing[1:,1]).astype(np.int64)
t_after = np.array(timing[1:,2]).astype(np.int64)
n_spans = np.array(timing[1:,3]).astype(np.int64)
t_diff = np.subtract(t_after, t_before)
t_diff_sorted = np.sort(t_diff)
```

```python
print("Num batches: " + str(len(batches.tolist())))
print("Num spans: " + str(np.sum(n_spans)))
```

```python
pyplot.plot(t_diff_sorted,"bo")
pyplot.xlabel('Popularity')
pyplot.ylabel('Latency deserialization (ms)')
pyplot.show()
```

```python
pyplot.hist(t_diff_sorted, 80, color='b')
pyplot.xlim(0,10)
pyplot.ylabel('Popularity')
pyplot.xlabel('Latency deserialization (ms)')
pyplot.show()
```

```python
pyplot.plot(t_diff, n_spans, "bo")
pyplot.ylabel('Number of spans')
pyplot.xlabel('Latency deserialization (ms)')
pyplot.show()
```

```python
pyplot.plot(t_before, t_diff, "go")
pyplot.xlabel('Time')
pyplot.ylabel('Latency serialization')
pyplot.show()
```

```python
pyplot.hist(t_before, 50, color='g')
pyplot.xlabel('Time arrival')
pyplot.ylabel('Count')
pyplot.show()
```
