```python
import matplotlib.pyplot as pyplot
import numpy as np
```

```python
timing  = np.genfromtxt('csv/thriftTiming.csv', delimiter=',', dtype='int')
batches = np.array(timing[1:,0])
t_before = np.array(timing[1:,1])
t_after = np.array(timing[1:,2])
n_spans = np.array(timing[1:,3])
t_diff = np.subtract(t_after, t_before)
t_diff_sorted = np.sort(t_diff)
```

```python
pyplot.plot(t_diff_sorted,"bo")
pyplot.xlabel('Popularity')
pyplot.ylabel('Latency deserialization')
pyplot.show()
```

```python
pyplot.hist(t_diff, 80, color='b')
pyplot.xlim(0,10)
pyplot.xlabel('Popularity')
pyplot.ylabel('Latency deserialization')
pyplot.show()
```

```python
pyplot.plot(t_diff, n_spans, "bo")
pyplot.ylabel('Number of spans')
pyplot.xlabel('Latency deserialization')
pyplot.show()
```

```python
pyplot.plot(t_before, t_diff, "go")
pyplot.xlabel('Time')
pyplot.ylabel('Latency serialization')
pyplot.show()
```

```python
pyplot.hist(t_before, 50, color='g')
pyplot.xlabel('Count')
pyplot.ylabel('Time arrival')
pyplot.show()
```
